```python
from rdkit import Chem
from rdkit.Chem import AllChem
from copy import deepcopy
from pisystem import PiSystem
```

```python
def show_2d_mol(mol, highlights=[]):
    """Show a 2D molecule"""
    mol_copy = deepcopy(mol)
    AllChem.Compute2DCoords(mol_copy)
    return Chem.Draw.MolToImage(mol_copy, size=(300, 300), highlightAtoms=highlights)
```

```python
smiles = "C12=C(C=CN2)C=CC=C1CC2=CC=CC=C2"
mol = Chem.MolFromSmiles(smiles)
mol = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol)
AllChem.UFFOptimizeMolecule(mol)
show_2d_mol(mol)
```

```python
ps = PiSystem(mol)
```

```python
print('Largest PI system size:', ps.pi_systems_size)
print('Largest PI system indices:', ps.largest_pi_system)
print(f'Largest Pi system area: {ps.area:.2f}' )
print(f'Largest Pi system length: {ps.max_distance:.2f}' )
```

```python
show_2d_mol(mol, highlights=ps.largest_pi_system)
```

```python
show_2d_mol(mol, highlights=ps.smallest_pi_system)
```
